## Key numerical results

The selected series is the AMOCatlas NAC `TRANS_NAC_PROXY` time series. The summary below reports record length, sampling interval, missing values, basic statistics, Welch settings, confidence interval metadata, and the Parseval variance check.

```python
for key in [
    "array",
    "variable",
    "time_start",
    "time_end",
    "dt_days",
    "welch_segment_length",
    "welch_overlap",
    "welch_degrees_of_freedom",
    "parseval_ratio",
    "dominant_timescale_days",
]:
    print(f"{key}: {summary[key]}")
print("stats:", summary["stats"])
```
